**Análisis exploratorio de datos- Dataset Ames housing dataset**

```python
# Cargar dataset
df = pd.read_csv("../Data/AmesHousing.csv")

# Vista general
df.shape

#El dataset cuenta con aproximadamente 3.000 observaciones y un alto número de variables,  esto indica que se debe realizar un análisis exploratorio detallado.
```

Análisis de la variable objetivo *SalePrice*

```python
# Estadísticas básicas
df["SalePrice"].describe()
```

```python
# Histograma
sns.histplot(df["SalePrice"])
plt.title("Distribución de SalePrice")
plt.show()
```

```python
# Boxplot
sns.boxplot(x=df["SalePrice"])
plt.title("Boxplot de SalePrice")
plt.show()
```

La variable objetivo presenta una asimetría positiva y presencia de valores extremos, esto es clave para plantear una transformación a la hora de crear el modelo.

**Separar variables numéricas y categóricas**

```python
# Variables numéricas y categóricas
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

len(num_cols), len(cat_cols)
```

```python
num_cols, cat_cols[:10]
```

El alto número de variables categóricas genera la necesidad de estrategias de codificación adecuadas, diferenciando entre variables nominales y ordinales.

**Valores faltantes**

```python
# Porcentaje de valores faltantes
missing = (
    df.isnull()
    .mean()
    .sort_values(ascending=False)
    .mul(100)
)

missing.head(15)
```

```python
# Variables con más del 30% de missing
missing[missing > 30]
```

En varias variables, los valores faltantes no representan ausencia de información sino ausencia de la característica (por ejemplo, piscina o callejón), lo cual debe tratarse cuidadosamente.

**Correlación de variables numéricas con el precio**

```python
# Correlación con SalePrice
corr = df[num_cols].corr()["SalePrice"].sort_values(ascending=False)

corr.head(10)
```

```python
# Heatmap de variables más correlacionadas
top_corr_cols = corr.index[:10]

sns.heatmap(
    df[top_corr_cols].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlación de variables numéricas con SalePrice")
plt.show()
```

Las variables relacionadas con el tamaño y la calidad de la vivienda muestran una alta correlación con el precio de venta.

**Relaciones clave**

```python
key_features = ["Gr Liv Area", "Total Bsmt SF", "Garage Area", "1st Flr SF"]

for col in key_features:
    plt.figure(figsize=(5,4))
    sns.regplot(x=df[col], y=df["SalePrice"])
    sns.scatterplot(x=df[col], y=df["SalePrice"])
    plt.title(f"{col} vs SalePrice")
    plt.show()
```

Se observa una relación aproximadamente lineal entre el área habitable y el precio, así como el área del primer piso y el precio, con presencia de outliers que deberán tratarse cuidadosamente.

**Variables categóricas importantes**

```python
# Boxplot de calidad general vs precio
sns.boxplot(x="Overall Qual", y="SalePrice", data=df)
plt.title("Overall Qual vs SalePrice")
plt.show()
```

El boxplot evidencia una relación positiva fuerte entre la calidad general de la vivienda y el precio de venta. A medida que aumenta Overall Qual, la mediana del SalePrice se incrementa de forma consistente, con una mayor dispersión en calidades altas, lo que implica la influencia de factores adicionales en viviendas premium.

### Conclusiones del Análisis Exploratorio

- La variable objetivo "SalePrice" presenta asimetría positiva, por lo que se considerará una transformación logarítmica.
- Existen numerosas variables categóricas, algunas de carácter ordinal.
- Los valores faltantes tienen un significado semántico importante en varias variables.
- Las características relacionadas con tamaño y calidad son las más influyentes en el precio.
